```python
# Create a data set ....
import numpy as np
import matplotlib.pyplot as plt


```

```python
NP=1000
# Create a sample time series
a = np.sin(np.linspace(1,100,num=NP))
b = np.random.choice([0,1],size=NP,p=[0.05,0.95])
```

```python
#np.sin(a)
c = np.zeros(NP)
idx=0
for i in range(NP) :
    c[i] = a[idx]
    if(b[i]==0):
        idx=0
    idx+=1
```

```python
plt.plot(c)
```

```python
### Datablock API
# https://docs.fast.ai/data_block.html
```

```python
from fastai.data_block import FloatList
```

```python
# ItemList class has all the goodies!

# Data must be gathered as tuples before loading into itemlist is seems
X=list(zip(a,b,c))
y=c
       
db = FloatList(items=X).add(FloatList(items=X)).\
    split_by_idx([700,1000])  # breaks into _.train, _.test

# .databunch()
```
